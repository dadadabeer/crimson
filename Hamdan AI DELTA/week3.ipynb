```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

n_samples = 100

TV = np.random.uniform(50,300,n_samples)
radio = np.random.uniform(10,100,n_samples)
social_media = np.random.uniform(5,50,n_samples)

sales = 5 + 0.8*(TV) +0.6*radio + 0.3*social_media + np.random.normal()
```

```python
print(sales)
```

```python
plt.scatter(sales, TV)
```

```python
data = pd.DataFrame({"TV": TV, "Radio": radio, "Social Media": social_media, "Sales": sales})

print(data)
```

```python
from sklearn.model_selection import train_test_split


x = data[["TV", "Radio", "Social Media"]]
y = data["Sales"]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)


```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
import pandas as pd

df = pd.read_csv("sample_data.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df_cleaned = df.dropna()
df_cleaned
```

```python
df['Age'] = df["Age"].fillna(df["Age"].mean())
df
```

```python
df['Salary'] = df["Salary"].fillna(df["Salary"].median())
df
```

```python
#drop duplicate rows

df.drop_duplicates()
```

```python
df= df.dropna()
df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[["Age", "Salary"]] = scaler.fit_transform(df[["Age", "Salary"]])

df
```

```python

#Label Encoding

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df["Gender"] = encoder.fit_transform(df["Gender"])

df
```

```python
#Onehotencoding

df = pd.get_dummies(df, columns=['Category'])

df
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

df
```

```python
# Example: Calculate Total Salary over 12 months
df['Total_Annual_Salary'] = df['Salary'] * 12

# Example: Create a binary feature based on a condition
df['Is_High_Salary'] = (df['Salary'] > 0.5).astype(int)

df
```
