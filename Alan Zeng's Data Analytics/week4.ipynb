```python
import random

#count heads and tails

slow = 0
medium = 9
fast = 0
total_flips = 10000
choices = ["slow", "medium", "fast"]

for _ in range(total_flips):
    flip = random.choices(choices, weights=[0.3, 0.5, 0.2])[0]
    # print(flip)
    if flip == "slow":
        slow += 1
    elif flip == "medium":
        medium += 1
    else:
        fast += 1

print(slow/total_flips)
print(medium/total_flips)
print(fast/total_flips)
```

```python
import random
import numpy as np

bike_rentals = [100, 102, 178, 32, 75, 4, 91, 202, 203]
means = []

for i in range(1000):
    sample = random.sample(bike_rentals, 9)
    sample_mean = np.mean(sample)
    means.append(sample_mean)




print(means)

```

```python
#Bootstrapping

import numpy as np
import matplotlib.pyplot as plt


bike_rentals = np.array([100, 102, 178, 32, 75, 4, 91, 202, 203, 125, 200, 143])

#  a new sample with exactly 12 of data points = [100, 100, 178, 32, 75, 4, 91, 203, 203, 125, 200, 143])

n_bootstrap = 10000
sample_size = len(bike_rentals)

all_means = []

for _ in range(n_bootstrap):
    bootstrap_sample = np.random.choice(bike_rentals, size=sample_size, replace=True)
    sample_mean = np.mean(bootstrap_sample)
    all_means.append(sample_mean)

print(all_means)

#i want 90% confidence interval, it means i want 90% of the middle data, cut the 5% at the start and at the end. meaning 5th percentile to 95th percentile



lower_bound = np.percentile(all_means, 5)
upper_bound = np.percentile(all_means, 95)

plt.hist(all_means, bins=30, color="blue", alpha=0.6, edgecolor="black")
plt.axvline(lower_bound, color="red", linestyle="dashed", label="5th Percentile")
plt.axvline(upper_bound, color="green", linestyle="dashed", label="95th Percentile")
plt.axvline(np.mean(all_means), color="black", linestyle="solid", label="Mean Estimate")
plt.xlabel("Mean of Bootstrap Sample")
plt.ylabel("Frequency")
plt.title("Bootstrap Mean Distribution")
plt.legend()
plt.show()
```

```python
print(lower_bound)
print(upper_bound)

print(lower_bound/2 + upper_bound/2)
```

The central limit theorem (CLT) states that the distribution of a sample will approximate a normal distribution (i.e., a bell curve) as the sample size becomes larger, regardless of the population's actual distribution shape.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random values from a normal distribution (mean=100, std=15)
data = np.random.normal(loc=100, scale=15, size=100000)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Simulated Normal Distribution")
plt.show()
```
