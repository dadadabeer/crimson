```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate sample 2D data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

plt.scatter(X[:, 0], X[:, 1])
plt.title("Unlabeled Data")
plt.show()

# Apply K-means
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot with cluster labels
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            s=200, c='red', marker='X', label='Centroids')
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Live.csv")

df.head(10)
```

```python
df.shape
```

```python
df.isnull().sum()
df.drop(['Column1', 'Column2', 'Column3', 'Column4'], axis=1, inplace=True)

df
```

```python
df.drop(['status_id', 'status_published'], axis=1, inplace=True)
df
```

```python
df['status_type'].unique()
```

```python
X = df
y = df["status_type"]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X["status_type"] = le.fit_transform(X["status_type"])
y = le.fit_transform(y)

X
```

```python
X["status_type"].unique()
```

```python
cols = X.columns

cols
```

```python
from sklearn.preprocessing import MinMaxScaler

ms = MinMaxScaler()
X = ms.fit_transform(X)

```

```python
X = pd.DataFrame(X, columns = [cols])
X
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)

labels = kmeans.labels_ #predicted labels
centroids = kmeans.cluster_centers_ #centroids locations


```

```python
correct_labels = sum(y == labels)

correct_labels/y.size
```

```python
kmeans.inertia_
```

```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2,random_state=0)

kmeans.fit(X)

labels = kmeans.labels_

# check how many of the samples were correctly labeled

correct_labels = sum(y == labels)

print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))

print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X, labels)

score
```

```python
#KNNS (regression or classification) supervised

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

# Load the data
iris = load_iris()
X, y = iris.data, iris.target

# Create a KNN model with k=3
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

# New flower to classify
new_flower = [[5.7, 2.8, 4.1, 1.3]]

# Predict class
pred = knn.predict(new_flower)
print("Predicted Class:", iris.target_names[pred[0]])
```
