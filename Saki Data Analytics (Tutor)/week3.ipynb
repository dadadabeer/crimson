```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Toyota.csv", index_col=0, na_values=["??", "????"])

df.head(20)

```

```python
df.dropna(axis=0, inplace=True)

df.head(10)
```

```python
#graph og price vs age

plt.scatter(x=df["Age"], y=df["Price"])
plt.title("Scatter plot of Price vs Age")
plt.xlabel("Age in Months")
plt.ylabel("Price (Euros)")
plt.show()
```

```python
#graph og km vs price

plt.scatter(x=df["KM"], y=df["Price"])
plt.title("Scatter plot of Price vs KM")
plt.xlabel("KM")
plt.ylabel("Price")
plt.show()
```

```python
#HISTOGRAM

plt.hist(x=df["Price"], bins=5, color="green", edgecolor="white")
plt.title("Histogram plot of PRICE")
plt.xlabel("PRICE")
plt.ylabel("FREQUENCY")
plt.show()
```

```python
def convert_doors(door):
    if door == "three":
        return 3
    elif door == "four":
        return 4
    elif door == "five":
        return 5
    else:
        return int(door)


df["Doors"] = df["Doors"].apply(convert_doors)

    


df.head(10)
```

```python
#create a function that takes in a price and returns budget if price is less than 15000, mid range if less than 20000, luxury otherwise

def price_category(price):
    if price < 15000:
        return ("Budget")
    elif price < 20000:
        return ("Mid-Range")
    else:
        return ("Luxury")


df["Price Category"] = df['Price'].apply(price_category)

df.head(10)
```

```python
def resale_value(row):
    depreciation_rate = 0.05  # Assume 5% depreciation per year
    return row["Price"] * (1 - depreciation_rate) ** row["Age"]

df["Resale_Value"] = df.apply(resale_value, axis=1)

df.head(20)
```
